```python
import os
import pandas as pd

# -----------------------------
# PATHS
# -----------------------------
BASE_DIR = os.path.dirname(os.path.abspath('/content/Laser Sandbox'))

INPUT_FOLDER = os.path.join(BASE_DIR, "/content/Laser Sandbox/ERP CSV")
OUTPUT_FOLDER = os.path.join(BASE_DIR, "/content/Laser Sandbox/Mazak Watch Folder")

INPUT_FILE = os.path.join(INPUT_FOLDER, "/content/Laser Sandbox/ERP CSV/Mock ERP.csv")
OUTPUT_FILE = os.path.join(OUTPUT_FOLDER, "orders_cleaned.csv")

# -----------------------------
# ENSURE FOLDERS EXIST
# -----------------------------
os.makedirs(INPUT_FOLDER, exist_ok=True)
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

# -----------------------------
# LOAD CSV
# -----------------------------
df = pd.read_csv("/content/Laser Sandbox/ERP CSV/Mock ERP.csv")

# -----------------------------
# CLEAN + SORT LOGIC
# -----------------------------
df["Order Qty"] = pd.to_numeric(df["Order Qty"], errors="coerce").fillna(0)
# Removed 'due_date' as it was not found in the DataFrame

df = df.sort_values(
    by=["Material Type", "Order Qty"],
    ascending=[True, True]
)

# -----------------------------
# WRITE OUTPUT
# -----------------------------
df.to_csv(OUTPUT_FILE, index=False)

print("Sandbox demo complete.")
print(f"Input read from: {INPUT_FOLDER}")
print(f"Output written to: {OUTPUT_FOLDER}")
```
